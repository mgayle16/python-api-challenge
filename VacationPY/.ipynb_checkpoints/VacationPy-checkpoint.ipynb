### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#lock in Lat & Lng locations
locations = vacation_df[["Lat","Lng"]]
humidity = vacation_df['Humidity']

fig = gmaps.figure()

#heat map overlay creation
heat_layer = gmaps.heatmap_layer(locations, weights = humidity, 
                                  dissipating=False, max_intensity=100,
                                  point_radius=1)

#implement heat map layer
fig.add_layer(heat_layer)

fig

#NOTE: Not sure why the map is not displaying, my key is correct, active and being called from correct file. 
# I tried trouble shooting via instructions below, google search, and worked with one of the TA's. The TA think
# it might be an issue with my google mpas API account
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = vacation_df
hotel_df["Hotel Name"]= ""
hotel_df
```

```python
params = {
            "radius": 5000,
          "type": "lodging",
          "keyword": "hotel",
          "key": g_key
}

for index, row in hotel_df.iterrows():
    lat=row["Lat"]
    lng=row["Lng"]
    params["location"]=f"{lat},{lng}"
    
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?" 
    
    
    response = requests.get(url, params=params)
    
    json_response = response.json()
    #pprint(json_response)
    
    try:
        hotel_df.loc[index, "Hotel Name"] = json_response["results"][0]["name"]
                                                    
    except (KeyError, IndexError):
        print("Skipped Location")
                                                               
hotel_df

#Note: my for loop looks good, but not sure why it's not finding hotel names, in your feedback if you can provide some clarity that would be great. 
```

```python
#remove rows with missing hotel names
blank_value = float("NaN")
hotel_df["Hotel Name"].replace("",blank_value,inplace=True)
hotel_df.dropna(subset = ["Hotel Name"], inplace=True)
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
marker_layer =  gmaps.marker_layer(locations,info_box_content=hotel_info)
fig.add_layer(marker_layer)

# Display figure
fig
#NOTE: Not sure why the map is not displaying, my key is correct, active and being called from correct file. 
# I tried trouble shooting via instructions below, google search, and worked with one of the TA's. The TA think
# it might be an issue with my google mpas API account
```
